```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("../imgs/cube.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

sift= cv2.SIFT_create()
kp= sift.detect(gray)
img= cv2.drawKeypoints(img, kp, img)
plt.imshow(img[...,::-1])
```

```python
import cv2
import numpy as np

img = cv2.imread("../imgs/cube.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

sift= cv2.SIFT_create()
kp= sift.detect(gray, None)
img= cv2.drawKeypoints(img, kp, None, flags=cv2.DrawMatchesFlags_DRAW_RICH_KEYPOINTS)
plt.imshow(img[...,::-1])
```
