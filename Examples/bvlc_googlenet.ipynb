```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

rows= open('../xmltxt/synset_words.txt').read().strip().split('\n')
s=[i.split()[1] for i in rows]
net= cv2.dnn.readNetFromCaffe('../xmltxt/bvlc_googlenet.prototxt', '../xmltxt/bvlc_googlenet.caffemodel')
img=cv2.imread('../imgs/lion.jpg')
blob= cv2.dnn.blobFromImage(img, 1, (244,244), (104, 117, 123))
net.setInput(blob)
pre= net.forward()
plt.imshow(img[...,::-1]);plt.title(s[np.argmax(pre)])
```
