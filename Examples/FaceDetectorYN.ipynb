```python
# Example 1 with image 

import numpy as np
import matplotlib.pyplot as plt
import cv2

img= cv2.imread('../imgs/peops.jpg')
model=cv2.FaceDetectorYN_create(r"D:\Git_Hub\vision\xml\face_detection_yunet_2023mar.onnx",'',(320, 320),0.9, 0.3, 5000)
h,w,_=img.shape
model.setInputSize((w, h))
res= model.detect(img)
if res[1] is not None:
    print(f"Number of Face: {len(res[1])}")

if res[1] is not None:
    for i in res[1]:
        c=i[:-1].astype(np.int32)
        cv2.rectangle(img,(c[0], c[1]), ((c[0]+c[2]),(c[1]+c[3])), (0, 255, 0), 4) # detect face
        cv2.circle(img, (c[4], c[5]), 4, (255, 255, 0), -1) # detect right eye
        cv2.circle(img, (c[6], c[7]), 4, (255, 255, 0), -1) # detect left eye
        cv2.circle(img, (c[8], c[9]), 4, (255, 255, 0), -1) # detect nose
        cv2.circle(img, (c[10], c[11]), 4, (0, 0, 255), -1) # detect right lip
        cv2.circle(img, (c[12], c[13]), 4, (0, 0, 255), -1) # detect left lip 
    cv2.imshow('faces', img)
    cv2.imwrite("saved_image.jpg", img)

    cv2.waitKey(0)


cv2.destroyAllWindows()  
```

```python
# Example 3 with camera and crop

import numpy as np
import matplotlib.pyplot as plt
import cv2

cap = cv2.VideoCapture(1)
model = cv2.FaceDetectorYN_create(r"D:\Git_Hub\vision\xml\face_detection_yunet_2023mar.onnx",'',(320, 320),0.9,0.3,5000)


while True:
    ret, frame = cap.read()
    h, w, _ = frame.shape
    model.setInputSize((w, h))
    res = model.detect(frame)
    if res[1] is not None:
        for i in res[1]:
            c = i[:-1].astype(np.int32)  # coordinate
            cv2.rectangle(frame,(c[0], c[1]),((c[0] + c[2]), (c[1] + c[3])),(0, 255, 0),4)  # detect face
            crop = frame[min(c[1], c[1] + c[3]): max(c[1], c[1] + c[3]),min(c[0], c[0] + c[2]): max(c[0], c[0] + c[2])]
    cv2.imshow('camera', frame)        
    if crop is not None:
        cv2.imshow('face', crop)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(1)
model = cv2.FaceDetectorYN_create("../xmltxt/face_detection_yunet_2023mar.onnx","",(320, 320),0.9, 0.3, 5000)


eye_cascade = cv2.CascadeClassifier("../xmltxt/haarcascade_eye.xml")

while True:
    ret, frame = cap.read()
    if not ret:
        break
    h, w = frame.shape[:2]
    model.setInputSize((w, h))
    _, faces = model.detect(frame)

    if faces is not None and len(faces) > 0:
        for face in faces:
            face = face.astype(int)
            x, y, bw, bh = face[:4]

            x1 = max(0, x - 20)
            y1 = max(0, y - 20)
            x2 = min(w, x + bw + 20)
            y2 = min(h, y + bh + 20)

        face_crop = frame[y1:y2, x1:x2].copy()
        face_gray = cv2.cvtColor(face_crop, cv2.COLOR_BGR2GRAY)
       
        eyes = eye_cascade.detectMultiScale(face_crop)

        for (ex, ey, ew, eh) in eyes[:2]:  
            cv2.rectangle(face_crop, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)
            cx, cy = ex + ew // 2, ey + eh // 2
            cv2.circle(face_crop, (cx, cy), 3, (0, 0, 255), -1)
    cv2.imshow("face", face_crop)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Example 2 --> two face 
import matplotlib.pyplot as plt
model = cv2.FaceDetectorYN_create("../xmltxt/face_detection_yunet_2023mar.onnx","",(320, 320),0.9, 0.3, 5000)
reg = cv2.FaceRecognizerSF.create('../xmltxt/face_recognition_sface_2021dec.onnx',"")
img1=cv2.imread('../imgs/a1.png')
h,w,_=img1.shape
model.setInputSize((w, h))
face1= model.detect(img1)
face_fe=reg.alignCrop(img1, face1[1][0])
face_fe=reg.feature(face_fe)

img2=cv2.imread('../imgs/a2.jpg')
h,w,_=img2.shape
model.setInputSize((w, h))
face2= model.detect(img2)
face_fe2=reg.alignCrop(img2, face2[1][0])
face_fe2=reg.feature(face_fe2)

l2=reg.match(face_fe, face_fe2, cv2.FaceRecognizerSF_FR_NORM_L2)
if l2<=1.0128:
    print('yes')
else:
    print('no')

plt.figure(figsize=(8, 2))
plt.subplot(121); plt.imshow(img1[..., ::-1]); plt.title("img1"); plt.axis("off")
plt.subplot(122); plt.imshow(img2[..., ::-1]); plt.title("img2"); plt.axis("off")
plt.show()    
```

```python
import cv2
import os
import numpy as np

model = cv2.FaceDetectorYN_create("../xmltxt/face_detection_yunet_2023mar.onnx","",(320, 320),0.6, 0.3, 5000)
reg = cv2.FaceRecognizerSF.create('../xmltxt/face_recognition_sface_2021dec.onnx',"")
path = r"../imgs/men"
data = os.listdir(path)
cam = cv2.VideoCapture(1)
check = []

while True:
    s = []
    nameimage = []
    ret, frame = cam.read()
    h, w, _ = frame.shape
    model.setInputSize((w, h))
    face1 = model.detect(frame)
    
    if face1[1] is not None:
        face_fe = reg.alignCrop(frame, face1[1][0])
        face_fe = reg.feature(face_fe)
        
        for i in data:
            x = os.path.join(path, i)
            img2 = cv2.imread(x)
            h, w, _ = img2.shape
            model.setInputSize((w, h))
            face2 = model.detect(img2)
            
            if face2[1] is not None:
                face_fe2 = reg.alignCrop(img2, face2[1][0])
                face_fe2 = reg.feature(face_fe2)
                score = reg.match(face_fe, face_fe2, cv2.FaceRecognizerSF_FR_NORM_L2)
                if score <= 1.0128:
                    s.append(score)
                    nameimage.append(i)
        
        if s != []:
            index = np.argmin(s)
            img_p = nameimage[index]
            name = img_p.split('.')[0]
            x = os.path.join(path, img_p)
            img = cv2.imread(x)
            cv2.imshow(name, img)
            check.append(name)
        
        if len(check) >= 2:
            if check[-1] != check[-2]:
                cv2.destroyWindow(check[-2])
                check.pop(0)
        
        cv2.imshow('Camera', frame)
        
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import os

model = cv2.FaceDetectorYN_create("../xmltxt/face_detection_yunet_2023mar.onnx","",(320, 320),0.9, 0.3, 5000)
reg = cv2.FaceRecognizerSF.create('../xmltxt/face_recognition_sface_2021dec.onnx',"")

path = r"../imgs/men"
data = os.listdir(path)
check= []

database = {}
for i in data:
    x = os.path.join(path, i)
    img2 = cv2.imread(x)
    if img2 is None:
        continue

    h, w = img2.shape[:2]
    model.setInputSize((w, h))
    _, face2 = model.detect(img2)

    if face2 is not None:
        aligned2 = reg.alignCrop(img2, face2[0])
        feature2 = reg.feature(aligned2)
        database[i] = feature2

print("Loaded:", list(database.keys()))


cam = cv2.VideoCapture(1)

while True:
    s = []
    nameimage = []

    ret, frame = cam.read()
    if not ret:
        break

    h, w,_ = frame.shape
    model.setInputSize((w, h))
    _, faces = model.detect(frame)

    if faces is not None:
        face_fe=  reg.alignCrop(frame, faces[0])
        face_fe = reg.feature(face_fe)

        for name, feature2 in database.items():
            l2 = reg.match(face_fe, feature2,
                           cv2.FaceRecognizerSF_FR_NORM_L2)

            if l2 <= 1.0128:
                s.append(l2)
                nameimage.append(name)

    if s:
        index = np.argmin(s)
        img_p = nameimage[index]
        name = img_p.split('.')[0]

        x = os.path.join(path, img_p)
        img = cv2.imread(x)
        cv2.imshow(name, img)

    cv2.imshow('Camera', frame)

    if cv2.waitKey(1) == ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```
